# Imputing null values

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values =np.nan , strategy ='mean' )
imputed_data=imputer.fit_transform(titanic[['Age']])
titanic['Age_modified']=pd.DataFrame( imputed_data).astype(int)
titanic.head(10)
```

```python
Age = titanic['Age']
titanic['Age'] = titanic['Age_modified']
del titanic['Age_modified']
```
